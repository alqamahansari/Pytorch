**1. Creating tensors**

```python
import torch

# tensors
t1 = torch.tensor([1, 2, 3])
print(t1, t1.shape)
```

```python
t2 = torch.tensor([[1, 2, 3], [4, 5, 6]])
print(t2, t2.shape)
```

```python
z = torch.zeros([2, 3])
print(z, z.shape)
```

```python
o = torch.ones([2, 3])
print(o, o.shape)
```

```python
r = torch.randn([2, 3])
print(r, r.shape)
```

```python
a1 = torch.arange(0, 10, 2)
print(a1, a1.shape)
```

```python
a2 = torch.linspace(0., 1., steps=5)
print(a2, a2.shape)
```

**4. Reshaping and dimension ops**

```python
x = torch.arange(12)
print(x, x.shape)
```

```python
x2 = x.reshape(3, 4)
print(x2, x2.shape)
```

```python
x3 = x.view(2, 6)
print(x3.shape)
```

```python
# Add dimensions
x4 = x.unsqueeze(0)    # shape: (1, 12)
print(x4.shape)
```

```python
x5 = x.unsqueeze(1)    # shape: (12, 1)
print(x5.shape)
```

```python
# Remove dimensions of size 1
x6 = x5.squeeze()
print(x6.shape)
```

```python
# 3D example
t3 = torch.arange(24).reshape(2, 3, 4)
print(t3.shape)
```

```python
# Swap dimensions (batch, channels, length) -> (batch, length, channels)
t3_perm = t3.permute(0, 2, 1)
print(t3_perm.shape)
```

**5. Elementwise ops vs matrix multiplication**

```python
A = torch.arange(6).reshape(2, 3).float()
B = torch.ones((2, 3))
```

```python
print("A:", A, A.shape)
print("B:", B, B.shape)
```

```python
# Elementwise
C = A + B
print("C:", C.shape)
```

```python
D = A * B
print("D:", D.shape)
```

```python
# Matrix multiplication: (2,3) x (3,4) -> (2,4)
M1 = torch.randn(2, 3)
M2 = torch.randn(3, 4)
M3 = torch.matmul(M1, M2)
print(M1.shape, M2.shape, M3.shape)
```

```python
# Alternative syntax
M4 = M1 @ M2
print(M4.shape)
```

**6. Moving tensors to GPU (conceptual habit)**

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
print("Device:", device)
```

```python
x = torch.randn(3, 4)
print("Before:", x.device, x.shape)
```

```python
x = x.to(device)
print("After:", x.device, x.shape)
```

```python
# Make sure operations keep device consistent
y = torch.randn(3, 4, device=device)
z = x + y
print(z.device, z.shape)
```
